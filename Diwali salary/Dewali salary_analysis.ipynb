```python
#dropping the column unnamed1 and status
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#delete the null values in Amount,since it has 12 null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#changing the data type in this case we choose the amount fro. float to int
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
#describe specific columns 
df[['Age', 'Orders', 'Amount']].describe()
```

```python
#Gender...in seaborn
ax = sns.countplot(x='Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#groupby based on the gernder and sort according to Amount
sale_gen = df.groupby(['Gender'], as_index = False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot (x = 'Gender', y = 'Amount', data = sale_gen)
```

```python
#count based on the age group and male /female using hue
ax= sns.countplot(x='Age Group', hue='Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group
sales_age = df.groupby(['Age Group'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x='Age Group',y ='Amount', data = sales_age)
```

```python
#Total number of order from top 10 states
Total_orders = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y='Orders', data = Total_orders)
```

```python
#marital status
sns.set(rc={'figure.figsize':(4,1.5
                             )})
ax = sns.countplot(x='Marital_Status', data = df)


#for lebel
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
#marital status vs amount
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Marital_Status', y = 'Amount', hue='Gender', data = sales_state)
```

```python
#occupation 
sns.set(rc={'figure.figsize':(20,4)})
ax=sns.countplot(x='Occupation', data = df)

#To see the value lebel
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Which occupation person can highest buying capacity.....occupation vs Amount
occupations=df.groupby(['Occupation', 'Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(x='Occupation', y= 'Amount', data = occupations, hue='Gender')
```

```python
#Which occupation person can highest buying capacity.....occupation vs Amount
occupations=df.groupby(['Occupation'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

ax = sns.barplot(x='Occupation', y= 'Amount', data = occupations)

#for leveling the data

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Product_Category vs amount
sns.set(rc={'figure.figsize':(25,5)})
sns.countplot(x='Product_Category', data = df)
```

```python
product_catagory = df.groupby(['Product_Category'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending = False).head (10)
                                                                                                                                      

sns.barplot(x= 'Product_Category', y = 'Amount', data = product_catagory)
```
